```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
# Creating the figure
plt.figure(figsize=(16, 16))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(df["Township"], df["Male"], label="Male")
plt.plot(df["Township"], df["Female"],"--", label="Female")
plt.plot(df["Township"], df["Total(Male+Female)"], "g--", label="Total(Male+Female)")
plt.ylabel('Number of people', color='black')
plt.xlabel('Township', color='black')
plt.title('The line graph of population')
plt.legend()
plt.xticks(rotation=40)

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(df["Township"], df["Immigration"], label="Immigration")
plt.plot(df["Township"], df["Emigration"],"--", label="Emigration")
plt.ylabel('Number of people', color='black')
plt.xlabel('Township', color='black')
plt.title('The line graph of Immigration and Emigration')
plt.legend()
plt.xticks(rotation=40)

# Subplot 3
plt.subplot(2, 2, 3)
plt.bar(df["Township"], df["Number_of_Households"])
plt.ylabel('Number of people', color='black')
plt.xlabel('Township', color='black')
plt.title('The bar graph of household number')
plt.xticks(rotation=40)

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(df["Township"], df["Birth"], label="Birth")
plt.plot(df["Township"], df["Death"],"--", label="Death")
plt.ylabel('Number of people', color='black')
plt.xlabel('Township', color='black')
plt.title('The line graph of births and deaths')
plt.legend()

# Adjusting the layout
plt.tight_layout()
plt.xticks(rotation=40)

# Setting title for the figure
plt.suptitle('Miaoli County population data', fontsize=18, fontweight='bold', y=1.02)

# saving the plot
plt.savefig('Miaoli_county_2.png', facecolor='white') 
# Displaying the plot
plt.show()
```
